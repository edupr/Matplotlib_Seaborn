```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

Correlacion (va de -1 a 1):
-1: no relación
0: No hay correlación
1: correlación directa

```python
tips.corr()
```

Con una gráfica heatmap es más fácil visualizar correlaciones

```python
sns.heatmap(tips.corr())
plt.show()
```

annot: valor  cmap: color map  

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm',linewidths=5,linecolor='black',
            vmin=0.5,vmax=1,cbar=True)
plt.show()
```
