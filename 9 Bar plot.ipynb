```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA','JAPAN','MEXICO','COLOMBIA','GERMANY']
population = [1000,800,900,1000,300]
```

```python
plt.bar(country,population,width=0.5,color=['gray','navy'])
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'),rotation=45)
plt.xlabel('País')
plt.ylabel('Millones de habitantes')
plt.title('Población Mundial')
plt.show()
```

```python
plt.barh(country,population)
plt.show()
```
