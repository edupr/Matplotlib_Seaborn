```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

## Histograma
bins: número de barras

```python
plt.hist(data,bins=10,histtype='bar')
plt.show()
```

```python
plt.hist(data,bins=10,histtype='step')
plt.show()
```

## Boxplot
vert: vertical
patch_artist:rellena el rango intercuartil
notch: enfoca datos hacia la mediana
showfliers: el 200 (outlier) lo grafica

```python
data = np.append(data,200)
plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=True)
plt.show()
```

### Scatter

```python
N = 50 
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 *  np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y,s=colors)
```

```python
plt.scatter(x,y,s=area)
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors)
```

```python
plt.scatter(x,y,s=area,c=colors,marker='o',alpha=0.5)
plt.show()
```
