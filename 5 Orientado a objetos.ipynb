```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

Los primeros 2 números es la posición. Los segundos 2 números son el tamaño

```python
fig = plt.figure()   # Creamos una figura (canvas vacío) 
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.1,0.1,0.4,0.3])
axes.plot(x,y,'b')
fig.show()
```

```python
fig = plt.figure()   # Creamos una figura (canvas vacío) 
axes = fig.add_axes([0.1,0.1,0.5,0.9])   #  Añadimos los ejes
axes.plot(x,y,'b')     # 'b' hace referencia al color azul (blue)
fig.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.9]) # eje principal
axes2 = fig.add_axes([0.2,0.55,0.4,0.3]) # eje interior

# Figura principal axes
axes.plot(x,y,'b')

# Inserta figura axes2
axes2.plot(y,x,'r')
axes2.set_facecolor('gray')

fig.show()
```
